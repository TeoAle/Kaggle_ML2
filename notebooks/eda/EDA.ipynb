```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
from ydata_profiling import ProfileReport
import plotly.express as px

sys.path.append('../../src')
from dataloader import *

%reload_ext autoreload
%autoreload 2

plots_theme = "plotly_dark"
```

```python
df = load_train_df(
    PATH = '../../data',
    decode_dummies=False
)
```

Converting horizontal and vertical distances to hydrology to the Euclidean one

```python
df['Distance_To_Hydrology'] = np.sqrt(df['Horizontal_Distance_To_Hydrology']**2 + df['Vertical_Distance_To_Hydrology']**2)

hydrology = df[['Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology', 'Distance_To_Hydrology']]
df.drop(['Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology'], axis=1, inplace=True)
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=1, cols=3, subplot_titles=("Horizontal Distance", "Vertical Distance", "Distance to Hydrology"))


fig.add_trace(
    go.Histogram(x=hydrology['Horizontal_Distance_To_Hydrology'], name='Horizontal Distance'),
    row=1, col=1
)

fig.add_trace(
    go.Histogram(x=hydrology['Vertical_Distance_To_Hydrology'], name='Vertical Distance'),
    row=1, col=2
)

fig.add_trace(
    go.Histogram(x=hydrology['Distance_To_Hydrology'], name='Distance to Hydrology'),
    row=1, col=3
)


fig.update_layout(
    height=400,
    width=1000,
    title_text="Distance to Hydrology",
    template=plots_theme
    )
```

```python
fig = make_subplots(rows=1, cols=3, subplot_titles=("Horizontal Distance", "Vertical Distance", "Distance to Hydrology"))


fig.add_trace(
    go.Histogram(x=np.log(hydrology['Horizontal_Distance_To_Hydrology'] + 1.), name='Horizontal Distance'),
    row=1, col=1
)

fig.add_trace(
    go.Histogram(x=np.log(hydrology['Vertical_Distance_To_Hydrology'] + 1.), name='Vertical Distance'),
    row=1, col=2
)

fig.add_trace(
    go.Histogram(x=np.log(hydrology['Distance_To_Hydrology'] + 1.), name='Distance to Hydrology'),
    row=1, col=3
)

fig.update_layout(
    height=400,
    width=1000,
    title_text="Distance to Hydrology",
    template=plots_theme
    )
```

```python
# barplots
distance_0 = df.loc[hydrology[hydrology.Distance_To_Hydrology <= 10].index].Cover_Type.value_counts()
vdistance_0 = df.loc[hydrology[hydrology.Vertical_Distance_To_Hydrology <= 10].index].Cover_Type.value_counts()
hdistance_0 = df.loc[hydrology[hydrology.Horizontal_Distance_To_Hydrology <= 10].index].Cover_Type.value_counts()

fig = make_subplots(rows=1, cols=3, subplot_titles=("Horizontal Distance", "Vertical Distance", "Distance to Hydrology"))

fig.add_trace(
    go.Bar(x=distance_0.index, y=distance_0.values, name='Distance to Hydrology'),
    row=1, col=1
)

fig.add_trace(
    go.Bar(x=vdistance_0.index, y=vdistance_0.values, name='Vertical Distance'),
    row=1, col=2
)

fig.add_trace(
    go.Bar(x=hdistance_0.index, y=hdistance_0.values, name='Horizontal Distance'),
    row=1, col=3
)

fig.update_layout(
    height=400,
    width=1000,
    title_text="Distance to Hydrology",
    template=plots_theme
    )
```

```python
df = load_train_df(
    PATH = '../../data',
    decode_dummies=True,
    add_geo_features=True
)
```

```python
# plot cover type distribution by wilderness area
import plotly.express as px

fig = px.density_heatmap(
    df,
    x='Cover_Type',
    y='Wilderness_Area',
    title='Cover Type Distribution by Wilderness Area',
    labels={'Wilderness_Area': 'Wilderness Area', 'Cover_Type': 'Cover Type'},
    color_continuous_scale='Blues',
    histfunc='count',
    nbinsx=7,
    nbinsy=4,
    # aspect='auto'
)

fig.update_xaxes(tickvals=list(range(1, 8)))
fig.update_yaxes(tickvals=list(range(1, 5)))
fig.update_layout(
    height=400,
    width=1000,
    template=plots_theme
    )

fig.show()
```

```python
# plot cover type distribution by wilderness area 

df_viz = df.groupby(['Cover_Type', 'Wilderness_Area']).size().reset_index(name='count')
df_viz = df_viz.pivot(index='Cover_Type', columns='Wilderness_Area', values='count').reset_index()

fig = go.Figure(data=[
    go.Bar(name='Wilderness Area 1', x=df_viz['Cover_Type'], y=df_viz[1]),
    go.Bar(name='Wilderness Area 2', x=df_viz['Cover_Type'], y=df_viz[2]),
    go.Bar(name='Wilderness Area 3', x=df_viz['Cover_Type'], y=df_viz[3]),
    go.Bar(name='Wilderness Area 4', x=df_viz['Cover_Type'], y=df_viz[4])
])

fig.update_layout(
    barmode='stack',
    title='Cover Type Distribution by Wilderness Area',
    xaxis_title='Cover Type',
    yaxis_title='Count',
    template=plots_theme
)

fig.show()
```
