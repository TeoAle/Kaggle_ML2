```python
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import sys
import plotly.express as px
import plotly.graph_objects as go

sys.path.append('../../src')

from dataloader import *
```

```python
PATH = '../../data/'
df = load_train_df(
    PATH,
    decode_dummies=True,
    add_geo_features=False
    )
```

```python
y, X = df.Cover_Type, df.drop('Cover_Type', axis=1)

X = StandardScaler().fit_transform(X)
```

```python
tsne = TSNE(
    n_components=2,
    random_state=0,
    perplexity=30
    )

X_2d = tsne.fit_transform(X)
```

```python
fig = px.scatter(x=X_2d[:, 0], y=X_2d[:, 1], color=y)

fig.update_layout(
    title='t-SNE visualization of the dataset',
    xaxis_title='t-SNE 1',
    yaxis_title='t-SNE 2',
    legend_title='Cover Type',
    showlegend=True,
    template='plotly_dark'
)

fig.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.decomposition import FactorAnalysis

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

fig = px.scatter(x=X_pca[:, 0], y=X_pca[:, 1], color=y)

fig.update_layout(
    title='PCA visualization of the dataset',
    xaxis_title='PCA 1',
    yaxis_title='PCA 2',
    legend_title='Cover Type',
    showlegend=True,
    template='plotly_dark'
)

fig.show()
```

```python
factor = FactorAnalysis(n_components=2)
X_factor = factor.fit_transform(X)
```

```python
fig = px.scatter(x=X_factor[:, 0], y=X_factor[:, 1], color=y)

fig.update_layout(
    title='Factor Analysis visualization of the dataset',
    xaxis_title='Factor 1',
    yaxis_title='Factor 2',
    legend_title='Cover Type',
    showlegend=True,
    template='plotly_dark'
)

fig.show()
```
